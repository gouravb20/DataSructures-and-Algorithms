### P

* All the problems that are solvable in polynomial time

### NP

* All the problems that are solvable either in polynomial time or non-polynomial time (possible exponential time)
* These problems' solutions can be verified in polynomial time

### NP-Complete

* A subset of NP problems
* all the problems in NP can be reduced to it

**** If an NP-Complete problem can be solved in polynomial time, than all NP problems can be solved in polynomial time, therefore proving P == NP

### NP-Hard

* all the problems in NP can be reduced to it
